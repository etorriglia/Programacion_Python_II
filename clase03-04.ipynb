```python
#cargar el dataset de referencia airquality
import statsmodels.api as sm

# Cargar el conjunto de datos 'airquality'
airquality = sm.datasets.get_rdataset('airquality').data
print(airquality.head())
```

```python
#realizar un boxplot del dataset airquality
sns.boxplot(data=airquality)
```

```python
#realizar el boxplor de airquality con matplotlib
plt.boxplot(airquality['Ozone'].dropna())
```

```python
#realizar el boxplot de ozono con seaborjn, con un boxplot por mes
sns.boxplot(x='Month', y='Ozone', data=airquality)
```

```python

bp = airquality.boxplot(by='Month', column='Ozone',  return_type='both')
```

```python
#listar los outliers de temnperatura para cada mes del año
Q1 = airquality.groupby('Month')['Ozone'].quantile(0.25)
Q3 = airquality.groupby('Month')['Ozone'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
#itero por los meses del año los datos de temperatura y listo los outliers
for month in range(5, 10):
    print(f"Mes: {month}")
    print(airquality[(airquality['Month'] == month) & ((airquality['Ozone'] < lower_bound[month]) | (airquality['Ozone'] > upper_bound[month]))])
```

```python
print(bp)
for x in bp:
    for y in x[1]:
        print(y, x[1][y])

print(bp[0][1]["fliers"])
for f in bp[0][1]["fliers"]:
    print(f.get_ydata())
```

```python
#creo una lista de 10 números con valores faltantes
data = [1, 2, np.nan,  6, 7, np.nan, 9, 10]
x = [1,2,3,4,5,6,7,8]
#grafico los dtos
plt.scatter(x, data)
plt.show()
```

```python
#intrapolar vaslores faltantes
sdata = pd.Series(data)
sdata.interpolate(inplace=True)
#grafico los datos interpolados
plt.scatter(x, sdata)
```

```python
#completar los valores faltantes repitiendo el último valor presente
sdata = pd.Series(data)
#sdata.fillna(method='ffill', inplace=True)
sdata.interpolate(method="pad", limit=2, inplace=True)
#grafico los datos interpolados
plt.scatter(x, sdata)
```

```python
#completar los valores faltantes suavizando la curva
sdata = pd.Series(data)
#sdata.fillna(method='ffill', inplace=True)
sdata.interpolate(method="piecewise_polynomial", order=2, inplace=True)
#grafico los datos interpolados
plt.scatter(x, sdata)
```

```python
#completar los valores faltantes suavizando la curva
sdata = pd.Series(data)
#sdata.fillna(method='ffill', inplace=True)
sdata.interpolate(method="cubicspline", order=2, inplace=True)
#grafico los datos interpolados
plt.scatter(x, sdata)
```

```python
import sklearn
import sklearn.datasets
diabetes = sklearn.datasets.load_diabetes()
#convertir a pandas dataframe
diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
diabetes.head()
```

```python
diabetesi = diabetes.copy()
#aleatoriamente, poner 50 datos en nulo en diversas columnas y rengloes del dataframe
idx = diabetes.sample(100).index
for x in idx:
    col = np.random.randint(1, 10)
    diabetesi.loc[x, diabetesi.columns[col]] = np.nan
```

```python
#mostrar um mapa de cantidad de nulos por columnas de diabetes
sns.heatmap(diabetesi.isnull(), cbar=False)
```

```python
XO = diabetes.values
XI = diabetesi.values

from fancyimpute import KNN, NuclearNormMinimization, SoftImpute, BiScaler

# Use 3 nearest rows which have a feature to fill in each row's missing features
X_filled_knn = KNN(k=3).fit_transform(XI)


#computar en MSE de los valores imputados contra los valores originales usando los dataframes diabetes y diabetesi
missing_mask = diabetesi.isnull().values

knn_mse = ((XO[missing_mask] - X_filled_knn[missing_mask]) ** 2).mean()
print("knnImpute MSE: %f" % knn_mse)
```
